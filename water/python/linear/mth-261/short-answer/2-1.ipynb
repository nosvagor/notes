```python
# Question 1
# I mean, we could add each element manually, but who likes arithmatic?

A = Matrix([
            [3,-3,0],
            [-7,6,-7],
            [3,-5,-4]
])
B = Matrix([
            [-7,10,7],
            [1,2,9],
            [10,-2,-6]
])

print(np.array(A+B))
```

```python
# Question 3
# Still pretty easy, just multiplication of each element, boring tho!

A = Matrix([
            [7,10,-3],
            [9,9,6],
            [-3,5,5]
])

# l = lambda, the scalar
l = 4

print(np.array(A*l))
print('\nScalar multiplcation is commutative (most of the time)!\n')
print(np.array(l*A))
```

```python
# Question 4

# we don't have to use sympy 'Matrix', np.array([[],[],[]]) works as too.
# I think rref is part of the symp package, which does need 'Matrix()' version.

A = np.array([
              [-1,2,3],
              [-10,-4,-2],
              [-9,-10,3]
])

# transposing is easy, simply use .T! 
print('Orignal Matrix')
print(A) 
print('\nTranposed Matrix')
print(A.T)
print('\nBack to Orignal Matrix')
print(A.T.T) 
```
